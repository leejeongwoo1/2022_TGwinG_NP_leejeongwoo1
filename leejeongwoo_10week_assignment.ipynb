* 동아리 회원들의 정보를 저장하는 Class를 만들 것이다.
* Member 클래스의 코드는 수정하지 않는다.
* Python 기본 Data Collection Type인 List를 사용하지 않는다.
* 구현해야할 클래스인 Members의 명세는 아래와 같다
* Linked List를 사용해서 풀이해야 함, Linked List에 대한 설명은 이 블럭의 마지막 부분 [참고 자료] 에 있음

### ```Class Member```

### ```Class Members``` 
##### 멤버 변수
* ```head = Member('headData', 'headData', 'headData', 'headData', False)```
* ```number = 0```

##### 멤버 함수
* ```addMember(self, id, year, name, gender, registered)```
    * id: 학번 (str)
    * year: 가입년도 (str)
    * name: 이름 (str)
    * gender: 성별 (str)
    * registered: 회비 납부 여부 (boolean)
    <br><br>
* ```delMember(self, id)```
    * 입력 받은 학번과 일치하는 회원을 삭제한다
    * 일치하는 회원이 없거나 회원 수가 0 이라면 "can not find %s member" %id를 출력하고 함수를 종료한다
<br><br>
* ```editMember(self, id)```
    * 입력 받은 학번과 일치하는 회원의 정보를 수정하는 함수를 불러온다
    * 일치하는 회원이 없거나 회원 수가 0 이라면 "can not find %s member" %id를 출력하고 함수를 종료한다
<br><br>
* ```editMemberDetail(self)```
    * 회원의 정보를 출력하고, 다음 줄에 어떤 정보를 수정할지 물어보는 텍스트를 출력한 후 사용자로 부터 입력 받는다.
    * 대체할 정보를 입력 받은 후 저장한다
<br><br>
* ```isLen(self)```
    * 동아리의 회원 수를 반환한다
<br><br>
* ```newSeason(self)```
    * 새 학기를 맞아 회비 납부 여부를 초기화한다
    * 모든 회원의 회비 납부 여부를 False로 바꾼다
<br><br>
* ```infoMem(self, id)```
    * 입력 받은 학번과 일치하는 회원의 정보를 출력한다
<br><br>
* ```printMembers(self)```
    * 모든 회원의 정보를 아래와 같이 보기 좋게 출력한다

|id|year|name|gender|registered|
|:-:|:-:|:-:|:-:|:-:|
|2017103000|2017|홍길동|남|1|


# 참고 자료
[강의노트 18. 자료구조 - LinkedList (링크드 리스트)](https://wayhome25.github.io/cs/2017/04/17/cs-19/) - 초보몽키의 개발공부로그

````python
class Member:
    id = None
    name = None
    year = None
    gender = None
    registered = None
    next = None

    def __init__(self, id, year, name, gender, registered):
        self.id = id
        self.name = name
        self.year = year
        self.gender = gender
        self.registered = registered
    def getInfo(self):
        return self.id, self.year, self.name, self.gender, self.registered
````

````python
class Members:
    head = Member('id', 'year', 'name', 'gender', 'registered')
    number = 0
    
    def addMember(self, id, year, name, gender, registered):
        n = self.head
        while True:
            if n.next == None:
                n.next = Member(id, year, name, gender, registered)
                break
                
            else:
                n = n.next
        self.number += 1      
                   
    def delMember(self, id):
        n = self.head
        while n.next != None:
            if n.next.id == id:
                t = n.next
                n.next = t.next
                del t
                break
            else:
                n = n.next
        if n.next == None:
            print("can not find %s member" %id)
            self.number+=1
        self.number -=1
            

    def editMember(self, id):
        n = self.head
        while n.next != None:
            if n.next.id == id:
                self.editMemberDetail(n.next)
                break
            else:
                n = n.next
        if n.next == None:
            print("can not find %s member" %id)
    def editMemberDetail(self, mem):
        q1 = input("어떤 정보 바꿀래요:")
        q2 = input("수정사항 입력:")
        if q1 =='name':
            mem.name = q2
        elif q1 =='id':
            mem.id = q2
        elif q1 =='year':
            mem.year = q2
        elif q1 =='gender':
            mem.gender = q2
        elif q1 =='registered':
            mem.registered = q2
    def isLen(self):
        return self.number
        
    def newSeason(self):
        n = self.head
        while n.next != None:
            n.next.registered = False
            n = n.next
    def infoMem(self, id):
        n = self.head
        while n.next != None:
            if n.next.id == id:
                print(n.next.id, n.next.year, n.next.name, n.next.gender, n.next.registered)
                break
            else:
                n = n.next
                
    def printMembers(self):
        n = self.head
        while n != None:
            print( '|',n.id,'|',n.year,'|', n.name,'|', n.gender,'|', n.registered,'|',sep='')
            n = n.next
````

## 아래의 테스트 코드가 문제 없이 동작하여야 함

````python
tgwing = Members()

tgwing.addMember('2017103000', '2017', '홍길동', '남', True)
tgwing.addMember('2018104000', '2018', '길동홍', '남', True)
tgwing.addMember('2019105000', '2019', '동홍길', '남', True)
tgwing.printMembers()
print("현재 회원 수:", tgwing.isLen())
print()

tgwing.delMember('2018104000')
tgwing.delMember('2018104000')
tgwing.printMembers()
print("현재 회원 수:", tgwing.isLen())
print()

tgwing.editMember('2019105000')
print()

tgwing.printMembers()
print()
tgwing.newSeason()

tgwing.infoMem('2017103000')
tgwing.printMembers()
````
